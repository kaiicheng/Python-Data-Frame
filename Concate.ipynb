```python
# 創造一個新result Data frame準備裝結果的資料
# result=pd.concat([data[0], data[1]], axis=1)
result=pd.DataFrame()
result
```

```python
for index, row in data.iteritems():
    if index == 0:
        pass
    else:
        print("Company: ", index)
        column_name=index
        company_data = data[column_name]
        date=data[0]
#         print(company_data)
#         print(date)

        # 合併日期欄及第一間公司欄位
        each_company_data=pd.concat([date, company_data], axis=1)
        each_company_data.rename(columns={index: 1}, inplace=True)  ########把Column名改成跟第一間公司的Column資料的Column一樣
        result=pd.concat([result, each_company_data], axis=0)  # 把這間公司的結果加入到總結果當中
        print(each_company_data)
        print(" ")
```
